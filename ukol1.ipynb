```python
elections = pandas.read_csv("1976-2020-president.csv")  #nacteni dat
elections
```

```python
elections["Rank"] = elections.groupby(["year", "state"])["candidatevotes"].rank(method = "min", ascending=False) #vytvoreni sloupce "Rank" a serazeni podle roku a statu,prirazeni hodnoty rank
elections.tail(15)
```

```python
elections_winners = elections[elections["Rank"] == 1.0] #vyselektovani vitezu
elections_winners
```

```python
elections_winners_sorted = elections_winners.sort_values(["state", "year"]) #serazeni podle roku a statu
elections_winners_sorted["previous_winner_party"]= elections_winners_sorted.groupby(["state"])["party_detailed"].shift(1)   #vytvoreni sloupce "previous winner party" a prirazeni hodnoty
elections_winners_sorted.head(15)
```

```python
elections_winners_sorted["change"] = numpy.where(elections_winners_sorted["previous_winner_party"] != elections_winners_sorted["party_simplified"], int(1),int(0))  #vytvoreni sloupce "change" a zadani podminky pro jeho obsah
elections_winners_sorted2 = elections_winners_sorted.loc[elections_winners_sorted["year"]!=1976]    #odstraneni roku 1976
elections_winners_sorted2.tail(15)
```

```python
data_pivot = elections_winners_sorted2.groupby(["state"])["change"].sum()   #serazeni dat a pocet hodnot sloupce "change"
data_pivot = pandas.DataFrame(data_pivot)
data_pivot = data_pivot.sort_values("change", ascending=False)
data_pivot.head(10)
```

```python
data = data_pivot.iloc[:10] #vytvoreni grafu pro 10 statu
data.plot(kind="bar")
```

```python
elections_winners_two = elections[elections["Rank"] < 2.1] #vygenerovani tabulky s vitezi voleb a druhymi v poradi
elections_winners_two
```

```python
elections_winners_votes = elections_winners_two.sort_values(["year", "state", "candidatevotes"], ascending= True)   #serazeni hodnot podle roku,statu a poctu hlasu
elections_winners_votes["margin"] = elections_winners_votes.groupby(["year", "state"])["candidatevotes"].shift(1)   #vytvoreni sloupce "margin" a dosazeni hodnot
elections_winners_votes["margin"] = abs(elections_winners_votes["candidatevotes"] - elections_winners_votes["margin"])  #absolutni hodnota vysledku odecteni hlasu viteze a druheho
elections_winners_margin = elections_winners_votes[elections_winners_votes["Rank"] == 1]    #vyselektovani pouze vitezove a o kolik hlasu vyhrali
elections_winners_margin.head()
```

```python
elections_winners_margin["relative_margin"] = elections_winners_margin["margin"]/elections_winners_margin["totalvotes"] #vytvoreni sloupce pro relativni hodnotu 
elections_winners_margin.head()
```

```python
elections_winners_relative_margin = elections_winners_margin.sort_values(["relative_margin"], ascending=True)
elections_winners_relative_margin.head()
```

```python
def swing(i):   #definice vlastni funkce pro prirazeni hodnot do sloupce "swing"
    if i["change"] == 1:
        if i["previous_winner_party"] == "DEMOCRAT":
            return "to Rep."
        else:
            return "to Dem."
    else:
        return "no swing"
    
elections_winners_sorted2["swing"] = elections_winners_sorted2.apply(swing,axis =1) #vytvoreni sloupce a pouziti funkce
elections_winners_sorted2.head()
```

```python
elections_pivot = pandas.pivot_table(data=elections_winners_sorted2,values="state", index ="year",columns = "swing", aggfunc="count")   #vytvoreni kontingencni tabulky - nektere roky mi tam o jedno nesedi s porovnanim dat z napovedy, ale nemohla jsem najit, kde je chyba
elections_pivot
```
